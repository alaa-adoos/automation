```python
import pandas as pd
import matplotlib.colors 
```

# Read the data

```python
df = pd.read_csv("daily_report_python.csv")
```

# split data

```python
df_1,df_2=df[:1], df[1:]
```

```python
df_2.columns=df_2.iloc[0]
df_2.drop(df_2.index[0],inplace=True)
df_2
```

# extract duration


```python
meeting_duration=df_1["Duration (Minutes)"][0]
meeting_duration = int(meeting_duration)
meeting_id = df_1["Meeting ID"][0]

meeting_duration
```

# get studant ID


```python
df_2['Student ID'] = df_2["User Email"].str.extract(pat='(^[0-9]*)')
```

# sum total of time for each studant


```python
df_2['Total']=df_2['Duration (Minutes)'].astype('int')
df_2=df_2.groupby('Student ID',as_index=False).sum(numeric_only=True)
df_2=df_2[1:]
df_2
```

# studant state


```python
df_2["state"]= ['Present' if x >= meeting_duration-10 else "Absent" if x <meeting_duration-20 else "late" for x in df_2['Total']] 
df_2['percentage of attendance']=[f'{round((x/meeting_duration)*100)}' for x in df_2['Total']]

df_2.style

df_2.style.background_gradient(subset=['percentage of attendance'], cmap='Greens')
```

```python
# Daily Report
```

```python
df_att=pd.read_csv('data.csv')
df_att[meeting_id]=df_2["state"]
df_att['Student ID']=df_2['Student ID']
df_att['percentage of attendance']=df_2['percentage of attendance']
df_att.to_csv('data.csv')
```
